```python
df = pd.read_csv(dados)
df = pd.DataFrame(df)
df.set_index(['Date'])
df.head()
```

```python
lista = []
for col in df.columns:
    lista.append(col.replace('Gauge: 0', '').replace('IMERG: ', ''))
lista.remove('Date')
cols = set(lista)
print(cols)
with open ('station_names.csv', 'w') as fp:
        fp.write(str(cols))
```

```python
def get_col_names(col):
    return f'Gauge: 0{col}', f'IMERG: {col}'
print(len(df))
def rbias(df,col):
    gauge, sat = get_col_names(col)
    gauge = df[gauge].dropna()
    sat = df[sat].dropna()
    num = sum(gauge-sat)*100
    den = sum(gauge)
    if den == 0:
        den = 0.000000001
    return (num/den)
```

```python
index = []
for col in cols:
    bias = rbias(df,col)
#    print(bias)
    index.append(bias)
print(index)
np.savetxt('bias.csv', index)
```
